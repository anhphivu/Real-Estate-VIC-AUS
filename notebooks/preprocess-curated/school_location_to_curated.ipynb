# School Data To Curated

Preprocess the raw school data to curated

## Import package

```python
import pandas as pd
```

```python
df = pd.read_csv('../../data/raw/raw_school_locations.csv')
df.head()
```

```python
df = df.rename(columns={'x_coor': 'longtitude', 'y_coor': 'latitude'})
df.head()
```

## Drop state and phone_number

```python
# Drop the 'state' and 'phone_number' columns
df = df.drop(columns=['state', 'phone_number', 'address'])
df.head()
```

```python
# Convert the suburb values to lowercase
df['suburb'] = df['suburb'].str.lower()
df.head()
```

## Save the data

```python
df.to_csv('../../data/curated/curated_school_location.csv', index = False)
```
